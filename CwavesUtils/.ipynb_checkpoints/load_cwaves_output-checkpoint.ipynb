### Plots
#### WF plots

```python
cluster_id = 151
pk_chan = 106 #get this information from clus_table.npy
n_chan_show = 16

fig, axs = plt.subplots(int(n_chan_show/2),2,figsize=(5,8), dpi=100, sharex=True, sharey=True)
for idx, (ch_idx, ax) in enumerate(zip(np.arange(pk_chan-int(n_chan_show/2),pk_chan+int(n_chan_show/2),1), axs.flat)):  
    ax.set_title('ch {}'.format(ch_idx))
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().set_ticks([])
    
    if ch_idx%2==1: 
        ax.spines['left'].set_visible(False)
        ax.yaxis.set_visible(False)
    if idx%4==0 or idx%4==1:
        ax_pos = ax.get_position()
        ax_pos.x0 = ax_pos.x0 + 0.05
        ax.set_position(ax_pos, which='original')
    
    sns.lineplot(ax=ax, x=range(82), y=mean_wf[cluster_id, ch_idx,:], lw=3)
```

Peak channel mean waveform plot

```python
cluster_id = 151
pk_chan = 106 #get this information from clus_table.npy

fig, ax = plt.subplots(1,1,figsize=(3,3), dpi=100)
ax.set_title('cluster {}'.format(cluster_id))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
sns.lineplot(ax=ax, x=range(82), y=mean_wf[cluster_id, pk_chan,:], lw=3)
ax.set_ylabel(r'$\mu$V')
```

#### SNR distribution plots

```python
fig, ax = plt.subplots(1,1,figsize=(3,3),dpi=200)
ax.hist(cluster_snr[:,0], bins=20, ec='w')
ax.set_xlim(0)
ax.set_xlabel('SNR')
ax.set_ylabel('Cluster count')

fig, ax = plt.subplots(1,1,figsize=(3,3),dpi=200)
ax.hist(cluster_snr[:,1], bins=20, ec='w')
ax.set_xlabel('Number of spikes in peak channel')
ax.set_ylabel('Cluster count')
```
